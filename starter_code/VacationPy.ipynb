### Humidity Heatmap
* Configure gmaps.
* Use the Lat and Lng as locations and Humidity as the weight.
* Add Heatmap layer to map.

```python
gmaps.configure(api_key=g_key)
locations = cities[["Latitude", "Longitude"]]
humidity = cities["Humidity (%)"].astype(float)

fig = gmaps.figure()
heat_layer = gmaps.heatmap_layer(locations, weights=humidity, 
                                 dissipating=False, max_intensity=10,
                                 point_radius = 1)
fig.add_layer(heat_layer)
fig
```

### Create new DataFrame fitting weather criteria
* Narrow down the cities to fit weather conditions.
* Drop any rows will null values.

```python
citiescopy = cities[['City','City ID','Latitude','Longitude','Humidity (%)','Temperature (F)','Wind Speed (mph)']].copy(deep=True)
ideal_cities = citiescopy.loc[(citiescopy["Temperature (F)"] < 90) & (citiescopy["Wind Speed (mph)"] < 10)]
ideal_cities
```

### Hotel Map
* Store into variable named `hotel_df`.
* Add a "Hotel Name" column to the DataFrame.
* Set parameters to search for hotels with 5000 meters.
* Hit the Google Places API for each city's coordinates.
* Store the first Hotel result into the DataFrame.
* Plot markers on top of the heatmap.

```python
url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
ideal_cities["Hotel"] = ''
params = {
    'type' : 'lodging',
    "radius" : 5000,
    'key' : g_key
}



for index, row in ideal_cities.iterrows():
    city = row['City']
    longitude = row['Longitude']
    latitude = row['Latitude']
    
    #params['keyword']=city
    params['location']=f'{latitude},{longitude}'
   
    
    hotels = requests.get(url, params=params).json()
    results = hotels['results']
    #print(hotels)
   
    
    try:
        ideal_cities.loc[index, 'Hotel'] = results[0]['name']
    except (KeyError, IndexError):
        print(f"Missing field for {city}, skipping")
    
    
```
